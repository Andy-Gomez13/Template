```python
#Arsenal (Ars) data worse start in years
Ars = pd.read_excel("/content/drive/MyDrive/Coding Dojo/Week 3/HW Items/Week51.xlsx")
Ars.info()
```

```python
numMatches = Ars.groupby(["Season"])["Position"].nunique()
numMatches.head()
```

```python
Matches = np.array(numMatches.index)
Matches
```

```python
#convert number items to readable list for graphing
Final_Pos = Ars["Position"].values
Final_Pos
```

```python
fig= plt.figure(figsize=(15,8))
plt.style.use("default")
plt.plot(Matches, Final_Pos, markersize = 10, c = 'r');
plt.ylim(bottom = 1, top =16);
plt.xlabel("Year", fontsize = 15);
plt.ylabel("Table Position", fontsize = 15);
plt.title("Arsenal's Table Position for Every Year");
plt.grid(axis = "y");
plt.grid(axis = "x");
plt.xticks(rotation = 90, fontsize = 10);
plt.yticks(fontsize = 10);
```
